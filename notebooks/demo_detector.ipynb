## Hierarchical Change-Point Detection

```python
import sys
import os
sys.path.append('..')

import numpy as np
from hiercpd.hiercpd import HierCPD # loading Hierarchical Detector package
```

### Categorical sequence

```python
Z = np.vstack((1*np.ones((20,1)), 2*np.ones((20,1)) ))
Z[1,:] = np.nan
Z[7,:] = np.nan
Z[15,:] = np.nan

# Sequence of ones and twos (K=2) with missing samples (NaN values). 
# This secuence will come from the previous module in a pipeline
# setting of the tool. I recommend to try with different toy sequences
# and increase the number of categories (K=10, K=20, ...)
```

### Model definition

```python
hyp_lambda = 1e3 # this hyper-parameter sets the inverse probability of change. (try other values)

cpd_model = HierCPD(Z, hyp_lambda)
cpd_model.detection()
cpd_model.plot_detection()
```
